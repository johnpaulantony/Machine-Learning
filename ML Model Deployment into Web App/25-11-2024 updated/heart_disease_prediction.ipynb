```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# load the dataset
data=pd.read_csv('heart.csv')
print(data.head())
```

```python
# Data preprocesing 

#1. Datacleaning - missing values, duplicate_entries, data type checking
#2. features scaling
#3. Encoder

# missing values
data.isna().sum()
```

```python
# duplicated entries
data.duplicated()
```

```python
# correct formatation
data.info()
```

```python
# Encode categorical features
print(list(data.columns))
categorical_features=[]
numerical_features=[]
for col in data.columns:
    if data[col].dtype == 'O':
        categorical_features.append(col)
    else:
        numerical_features.append(col)
categorical_features

encoders={col:LabelEncoder() for col in categorical_features}

for col in categorical_features:
    data[col]=encoders[col].fit_transform(data[col])
```

```python
data
```

```python
# Scale Numerical Features
scaler=StandardScaler()
numerical_features=numerical_features[0:-1]
numerical_features
data[numerical_features]=scaler.fit_transform(data[numerical_features])
```

```python
data
```

```python
# segregate and split the dataset
X=data.drop('HeartDisease',axis=1)
Y=data['HeartDisease']
```

```python
# split the dataset
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# train the naive bayes model
model=GaussianNB()
model.fit(X_train,Y_train)
```

```python
# Save model, encoders, scaler

import pickle
with open('model.pkl','wb') as f:
    pickle.dump(model,f)
with open('scaler.pkl','wb') as f:
    pickle.dump(scaler,f)
with open('encoders.pkl','wb') as f:
    pickle.dump(encoders,f)
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy=accuracy_score(Y_test,y_pred)
accuracy
```
