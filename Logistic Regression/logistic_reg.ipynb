```python
# segregating Dataset into X(features or input) & Y(output or label)
X=dataset.loc[:,'Age':'Salary'].values
X
```

```python
Y=dataset.loc[:,'Status'].values
Y
```

```python
# Splitting the dataset into Train and Test
# sklearn module - all the ML algorithms wriiten the packages
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
len(X_train)
len(X_test)
len(Y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
#select alg
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

```python
age=int(input("ENter the new customer age"))
salary=int(input("Enter the salary"))
newcust=[[age,salary]]
result=model.predict(sc.transform(newcust))
print(result)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
result_score=accuracy_score(Y_test,y_pred)
result_score
```
